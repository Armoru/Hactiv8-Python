```python
import pandas as pd
import seaborn as sbr
import matplotlib.pyplot as plt
```

```python
with open("book_new.csv") as csv_file:
    df_csv = pd.read_csv(csv_file,delimiter=";")
```

```python
len(df_csv.columns)
```

```python
type(df_csv)
```

```python
df_csv.drop(df_csv.columns[10], axis=1, inplace=True)
```

```python
df_csv
```

```python
df_csv['avg_rating'] = pd.to_numeric(df_csv['average_rating'],errors='coerce')
```

```python
df_csv = df_csv.dropna()
```

```python
group_author = df_csv[["authors","avg_rating"]].groupby(["authors"]).mean()
```

```python
count_avg= df_csv[['authors', 'avg_rating', 'bookID']].groupby(['authors']).agg(['count', 'mean'])
```

```python
count_avg
```

```python
highest_cm=count_avg.avg_rating.sort_values(by=['count','mean'],ascending=[False,False])
```

```python
high_cm=highest_cm.reset_index()
```

```python
h_cm=high_cm.head(5)
```

```python
h_cm
```

```python
h_cm.rename(columns = {'count':'Number_of_Books','mean':'Mean_Rating'}, inplace = True)
```

 ## FIRST CHART HIGHEST number book of release

```python
barplot = sbr.barplot(x='authors',y='Number_of_Books',data=h_cm)
barplot.set_title('Highest Book Release / Ratings')
```
