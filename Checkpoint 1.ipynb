```python
#find how many movies released in a year
year_dict={}
for data in list_data:
    #print (data['year']) #LIST
    if data['year'] is None:
        if 'Unknown' in year_dict.keys():
            year_dict['Unknown'] = year_dict['Unknown']+1
        else:
            year_dict['Unknown'] = 0
    if data['year'] in year_dict.keys():
        year_dict[data['year']] = year_dict[data['year']]+1
    else:
        year_dict[data['year']] = 1
```

```python
def drill_year(year):
    int_year = int(year)
    year_dict={}
    for data in list_data:
        #print (data['year']) #LIST
        if data['year'] is None:
            if 'Unknown' in year_dict.keys():
                year_dict['Unknown'] = year_dict['Unknown']+1
            else:
                year_dict['Unknown'] = 0
        if data['year'] in year_dict.keys():
            year_dict[data['year']] = year_dict[data['year']]+1
        else:
            year_dict[data['year']] = 0
    return year_dict[int_year]
```

```python
year = input('What year you want to see?\n')

try:
    print("Movies released in %s are %s movies" %(year,drill_year(year)))
except KeyError:
    print("Movies released in %s is not found"%(year))
```

```python
def drill_genre(genre):
    get_genre = str(genre)
    genre_dict={}
    for data in list_data:
        for g_data in data['genres']:
            if g_data =='':
                if 'Unknown' in genre_dict.keys():
                    genre_dict['Unknown'] = genre_dict['Unknown'] +1
                else:
                    genre_dict['Unknown'] = 0
            elif g_data in genre_dict.keys():
                genre_dict[g_data] = genre_dict[g_data] +1
            else:
                genre_dict[g_data] = 0
        return genre_dict[get_genre]
```

```python
genre= input('What genre you want to see?\n')
try:
    print("Movies released in %s are %s movies" %(genre,drill_genre(genre)))
except KeyError:
    print("Movies released in %s is not found"%(genre))
```
