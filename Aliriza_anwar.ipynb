```python
def drill_year(year):
    int_year = int(year)
    year_dict={}
    for data in list_data:
        #print (data['year']) #LIST
        if data['year'] is None:
            if 'Unknown' in year_dict.keys():
                year_dict['Unknown'] = year_dict['Unknown']+1
            else:
                year_dict['Unknown'] = 1
        elif data['year'] in year_dict.keys():
            year_dict[data['year']] = year_dict[data['year']]+1
        else:
            year_dict[data['year']] = 1
    return year_dict[int_year]
```

```python
year = input('What year you want to see?\n')

try:
    print("Movies released in %s are %s movies" %(year,drill_year(year)))
except KeyError:
    print("Movies released in %s is not found"%(year))
```

```python
def drill_genre(genre):
    get_genre = str(genre)
    genre_dict={}
    for data in list_data:
        if len(data['genres']) ==0:
            if 'Unknown' in genre_dict.keys():
                genre_dict['Unknown'] = genre_dict['Unknown'] +1
            else:
                genre_dict['Unknown'] = 1
        else:
            for g_data in data['genres']:
                if g_data in genre_dict.keys():
                    genre_dict[g_data] = genre_dict[g_data] +1
                else:
                    genre_dict[g_data] = 1
    return genre_dict[get_genre]
```

```python
genre= input('What genre you want to see?\n')
try:
    print("Movies released in %s are %s movies" %(genre,drill_genre(genre)))
except KeyError:
    print("Movies released in %s is not found"%(genre))
```

```python
#find how many cast
cast_dict={}
for data in list_data:
    if len(data['cast']) ==0:
        if 'Unknown' in cast_dict.keys():
            cast_dict['Unknown'] = cast_dict['Unknown'] +1
        else:
            cast_dict['Unknown'] = 1
    else:
        for g_data in data['cast']:         
            if g_data in cast_dict.keys():
                cast_dict[g_data] = cast_dict[g_data] +1
            else:
                cast_dict[g_data] = 1
```

```python
#find how max cast in  years
max_cast = 0
key_cast = ''
for data in cast_dict.keys():
    if data != 'Unknown':
        if max_cast < cast_dict[data]:
            max_cast = cast_dict[data]
            key_cast = data
print ("%s : %s"%(key_cast,max_cast)) #max cast : movie
#print(cast_dict.keys())
```
